```python
#Statistics Module
#The statistics module provides functions to mathematical statistics of numeric data. The following popular statistical functions are defined in this module.

#Mean
#The mean() method calculates the arithmetic mean of the numbers in a list.
import statistics
mean1=statistics.mean([2,5,6,9])
print("mean1:",mean1)


#Median
#The median() method returns the middle value of numeric data in a list.
import statistics
median1=(statistics.median([1,2,3,8,9]))
median2=(statistics.median([1,2,3,7,8,9]))
print("median1:",median1)
print("median2:",median2)


#Mode
#The mode() method returns the most common data point in the list.
import statistics
mode1=statistics.mode([2,5,3,2,8,3,9,4,2,5,6])
print("mode1:",mode1)


#Standard Deviation
#The stdev() method calculates the standard deviation on a given sample in the form of a list.

import statistics
SD=statistics.stdev([1,1.5,2,2.5,3,3.5,4,4.5,5])
print("Stadard Deviation:",SD)


#variance
#Return the sample variance of data, an iterable of at least two real-valued number
import statistics
var=(statistics.variance([1, 3, 5, 7, 9, 11]))
print("Variance:",var)
```
